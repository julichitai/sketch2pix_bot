```python
from selenium import webdriver
import time
from tqdm import tqdm_notebook
import requests
import matplotlib.pyplot as plt
from PIL import Image
import io
import numpy as np
from pathlib import Path
```

```python
driver = webdriver.Firefox(executable_path='../../Загрузки/geckodriver/geckodriver-v0.26.0-linux64/geckodriver')
url = 'https://yandex.ru/images/search?text=%D0%B3%D0%BE%D0%BB%D1%83%D0%B1%D1%8C&isize=small&type=photo&from=tabbar'
driver.get(url)
```

```python
# get full size
animal = 'pigeons'

path_to_data = Path(f'data/dataset/{animal}/original')
if not Path.exists(path_to_data):
    Path.mkdir(path_to_data)
        
images = driver.find_elements_by_class_name('serp-item__link')
name = 1
print(len(images), name)
for i in tqdm_notebook(images[:]):
    
    i.click()
    url = driver.find_element_by_class_name('MMImage_origin').get_attribute('src')
    
    if not url.endswith('.jpg'):
        url += '.jpg'
    try:
        img = requests.get(url).content

        image = Image.open(io.BytesIO(img))
        image = np.array(image)
#         plt.imshow(image)
#         plt.show()
        image = image.astype(np.float32)/255.0
        plt.imsave(path_to_data / Path(f'{animal}_{name}.jpg'), image)
        time.sleep(1.2)
        name += 1
    except Exception:
        pass
    
    driver.find_element_by_class_name('MMViewerModal-Close').click()
    
    if len(list(path_to_data.glob('*.jpg'))) > 260:
        break
```

```python
driver.find_element_by_xpath('//*[@id="text-input"]').clear()
driver.find_element_by_xpath('//*[@id="text-input"]').send_keys('привет')
driver.find_element_by_xpath('/html/body/div[1]/div[4]/div[2]/form/div/div[3]/div/button').click()
time.sleep(2)
ans = driver.find_element_by_xpath('/html/body/div[1]/div[4]/p').text.lower()
# answer = answer.append({'id': i, 'ans': ans})
ans
```

```python
import pandas as pd
test = pd.read_csv('test', sep=' ', header=None)
test = test.iloc[2700:2800]
answer = pd.DataFrame()

for i, row in tqdm_notebook(test.iterrows()):
#     print(i, row[1])
    driver.find_element_by_xpath('//*[@id="text-input"]').clear()
    driver.find_element_by_xpath('//*[@id="text-input"]').send_keys(row[1])
    driver.find_element_by_xpath('/html/body/div[1]/div[4]/div[2]/form/div/div[3]/div/button').click()
    time.sleep(0.11)
    ans = driver.find_element_by_xpath('/html/body/div[1]/div[4]/p').text.lower()
    answer = answer.append({'id': i, 'ans': ans}, ignore_index=True)
    print(i, ans)
answer
```

```python
answer['lang'] = answer.apply(lambda x: x.ans.split()[2], axis=1)
answer['id'] = answer['id'].astype('int64')
answer
```

```python
a['sub'] = list(lang_dict.values())
a['lang'] = list(lang_dict.keys())
a
```
